```python
sns.boxplot(x="Attrition", y="Age", data=df)
plt.title("Age vs Attrition")
plt.show()

sns.boxplot(x="Attrition", y="MonthlyIncome", data=df)
plt.title("MonthlyIncome vs Attrition")
plt.show()

sns.countplot(x="OverTime", hue="Attrition", data=df)
plt.title("OverTime vs Attrition")
plt.show()
```

```python
education_map = {
    1: 'Below College',
    2: 'College',
    3: 'Bachelor',
    4: 'Master',
    5: 'Doctor'
}

viz_data = df.copy()
viz_data['Education'] = viz_data['Education'].map(education_map)
 
```

```python
edu_attr_rate = viz_data.groupby("Education")["Attrition"].apply(lambda x: (x == "Yes").mean() * 100).sort_values()

# Plot
plt.figure(figsize=(8,5))
sns.barplot(x=edu_attr_rate.index, y=edu_attr_rate.values, palette="viridis")

plt.title("Attrition Rate by Education Level")
plt.ylabel("Attrition Rate (%)")
plt.xlabel("Education Level")
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
worklife_map = {
    1: 'Bad',
    2: 'Good',
    3: 'Better',
    4: 'Best'
}

viz_data = df.copy()
viz_data['WorkLifeBalance'] = viz_data['WorkLifeBalance'].map(worklife_map)
```
